### 1.4 Line/Scatter Graph

```python
# Make three lists of numbers
x = [1, 2, 3, 5, 6]
y1 = [1, 4.5, 7, 24, 38]
y2 = [1, 4, 9, 25, 36]

# (1.1) Make a 1st Scatter object
trace1 = Scatter(
    x=x,           # x-coordinates of trace
    y=y1,          # y-coordinates of trace
    mode='markers'   # scatter mode (more in UG section 1)
)

# (1.2) Make a 2nd Scatter object
trace2 = Scatter(
    x=x,           # same x-coordinates
    y=y2,          # different y-coordinates
    mode='lines'     # different scatter mode
) 

# (2) Make Data object 
data = Data([trace1, trace2])  # (!) Data is list-like, must use [ ]

# (3) Make Layout object (Layout is dict-like)
layout = Layout(title='Fig 0.3: Some Experiment')

# (4) Make Figure object (Figure is dict-like)
fig = Figure(data=data, layout=layout) 
```

```python
print(fig)  # print the figure object in notebook
```

```python
print(fig.to_string())  # better option - print figure object in human-friendly form
```

```python
# Send Figure object to Plotly and show plot in notebook
py.iplot(fig, filename='s0_second-plot') 
```

### 1.5 Adding Information

```python
# Import graph objects XAxis and YAxis (Call Help for more info)
from plotly.graph_objs import XAxis, YAxis
```

```python
# (6.1) Make XAxis object, add title key
xaxis = XAxis(title='Some independent variable')

# (6.2) Make YAxis object, add title key
yaxis = YAxis(title='Some dependent variable')

# (7) Update 'layout' key in the Figure object
fig['layout'].update(
    xaxis1=xaxis,  # link XAxis object to 'xaxis1' (corresp. to first/only x-axis)
    yaxis1=yaxis   # similarly for 'yaxis1'
)
```

```python
print(fig.to_string()) # print figure object to stdout
```

```python
# Send Figure object to Plotly and show plot in notebook
py.iplot(fig, filename='s0_second-plot-axis-titles') 
```

## 2. Line - Scatter

### 2.1 Changing Quartz Chart

```python
# Embed Plotly graph in notebook
#   Quartz is username of the graph's maker,
#   7 is the graph's unique id number
tls.embed('Quartz', '7')
```

```python
# Get figure object from Plotly's servers
Quartz7 = py.get_figure('Quartz', '7')
```

```python
print(Quartz7.to_string())  # print in notebook
```

```python
# Retrieve data from figure object (get a dictionary in return)
Quartz7_data = Quartz7.get_data()

Quartz7_data   # show in notebook
```

```python
Quartz7_data[0]
```

```python
# Make scatter trace object with data for China
trace_china = Scatter(
    Quartz7_data[0],  # (!) 'name', 'x' and 'y' as in original figure
    mode='markers+lines'      # marker pts and lines (the default here)
)
```

```python
print(trace_china.to_string())
```

```python
trace_usa = Scatter(
    Quartz7_data[1],  # (!) 'name', 'x' and 'y' as in original figure
    mode='markers',           # show marker pts only
    marker=Marker(
        symbol='square'       # show square marker pts
    )
)
```

```python
print(trace_usa.to_string())
```

```python
# Make data object containing two traces
data = Data([trace_china, trace_usa])
```

```python
# Make a layout object
layout = Layout(title='Fig 1: Number of university graduates [in millions]')
```

```python
# Make a figure object
fig = Figure(data=data, layout=layout)
```

```python
# Send figure object to Plotly and show result in notebook 
py.iplot(fig, filename='s1_grad-china-usa')
```
